```python
import pandas as pd
import glob
```

<p>All **.xls** files with metadata into one Data Frame</p>

```python
path = r'/home/taorkon/VicBinRum/metadata' # path to directory with metadata
all_files = glob.glob(path + "/*.xlsx")

li = []

for filename in all_files:
    df = pd.read_excel(filename, index_col=None, header=0)
    li.append(df)

metadata_df = pd.concat(li, axis=0, ignore_index=True, sort=False)
```
